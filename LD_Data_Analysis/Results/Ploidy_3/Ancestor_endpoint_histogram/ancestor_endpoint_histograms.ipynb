```python
import pandas as pd
```

```python
df_common = pd.read_csv ("common_histogram.dat", delimiter = " ")
df_common[0:10]
```

```python
df_endpoint = pd.read_csv ("endpoint_histogram.dat", delimiter = " ")
df_endpoint[0:10]
```

```python
import matplotlib.pyplot as plt

# plot results
fig = plt.figure(figsize=(11,7))
ax = fig.add_subplot(1, 1, 1)

### set x y range
ax.set_xlim(-0.0025, 0.2025)


#plt.title("Ruolo del numero di basi\n"+str(N)+" simulazioni ogni punto")
ax.set_ylabel('Conteggi (log)', fontsize=20)
ax.set_xlabel('Frequenza', fontsize=20)
ax.yaxis.offsetText.set_fontsize(20)

## set ticks
ax.tick_params(which='major', width=1.0, labelsize=20)
ax.tick_params(which='major', length=10, labelsize=20)

ax.plot(df_endpoint['x'], df_endpoint['counts'],
         c='tab:blue', lw=3,label='Punto finale')

ax.plot(df_common['x'], df_common['counts'],
         c='tab:orange', lw=3,label='Post confronto')

ax.legend(fontsize=20, title='Legenda',title_fontsize=20)
ax.set_yscale('log')
plt.tight_layout()
plt.savefig("ancestor_endpoint_comparison.pdf")
```
