```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl

mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['axes.labelsize'] = 14
mpl.rcParams['legend.fontsize'] = 12
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
```

```python
df = pd.read_csv ("parameters_results.csv")
print (df)
```

```python
df.plot("extant", "p_hat", marker = "o")
df.plot("extant", "min_mu", marker = "o")
df.plot("extant", "max_mu", marker = "o")
```

```python
p_hat = [x for x in df["p_hat"]]

extant = [x for x in df["extant"]]

min_mu = [x for x in df["min_mu"]]
max_mu = [x for x in df["max_mu"]]



fig = plt.figure(figsize=(8,8))

ax = plt.gca()
ax.plot(extant, p_hat, marker = "o", label = r'$\hat P$')
ax.grid()
ax.set_title(r'$\hatP$ for different $N_{extant}$')
ax.set_xlabel(r'$N_{extant}$')
ax.set_ylabel(r'$\hatP$')
ax.set_yscale('log')
ax.legend()
plt.savefig("p_hat.svg")
```

```python
mu = [0.5*(min_mu[i]+max_mu[i]) for i in range(len(min_mu))]
mu_ybar = [max_mu[i] - mu[i] for i in range(len(min_mu))]

print(mu)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.gca()
ax.errorbar(extant, mu, yerr=mu_ybar, capsize=8, marker = "o", label = r'$\hat \mu$')
ax.axhline(y=  1.0019844876411e-09, ls = "--", c='tab:red', label = 'MA')
#ax.set_yscale('log')

ax.grid()

ax.set_title(r'Estimated $\hat\mu$ range for different $N_{extant}$')
ax.set_xlabel(r'$N_{extant}$')
ax.set_ylabel(r'$\hat \mu$ range')
ax.legend()
ax.set_yscale("log")

plt.savefig("estimated_mu.svg")
```
